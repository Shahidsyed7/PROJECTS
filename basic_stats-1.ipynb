standardization of numerical variables
Z-Score normalization, is a technique used to scale numerical data so that it has a mean of 0 and a standard deviation of 1

```python
# Standardize the numerical columns
for column in df.select_dtypes(include=[np.number]): 
    df[column] = (df[column] - df[column].mean()) / df[column].std()
df
```

Conversion of Categorical Data into Dummy Variables
In machine learning, categorical data cannot be directly used by most algorithms as they expect numerical inputs. One-hot encoding converts categorical values into a numerical format that algorithms can process easily.

```python
category = ['Day', 'SKU', 'City', 'BU', 'Brand', 'Model']

for col in category:
    df[col] = df[col].astype(str)

# Apply one-hot encoding on categorical columns
df_encoded = pd.get_dummies(df[category], prefix=category, drop_first=False)
df_encoded = df_encoded.astype(int)

# Drop original categorical columns 
df = pd.concat([df.drop(columns=category), df_encoded], axis=1)
```
